## 결측값
1. 결측값 '-' 형태로 나타남
2. '-' 을 NaN으로 바꿔주기
3. 데이터 프레임을 구분항목별로 나눈뒤에 결측값 있는 열 drop 하기 

```python
#dt.isna().sum().sum() # 결측값 0으로 나옴
#dt[dt['2003']=='-'] #결측값 '-'형태로 저장되어 있음
dt.replace('-',np.nan,inplace=True)
```

# 행복지수 항목 컬럼 바꾸기

```python
dt.drop(index=0,inplace=True)
new_col = dt.columns.str.replace(".",'-').str.replace('-1','건강').str.replace('-2','재정').str.replace('-3','관계').str.replace('-4','가정').str.replace('-5','사회')
dt.columns = new_col
dt.drop(index=1)
dt.reset_index(drop=True,inplace=True)
dt.head()
```

## 구분 항목별로 데이터 프레임 나누기

```python
def rdr(dt):
    dt.reset_index(drop=True, inplace=True) #인덱스 재정렬
    dt.drop(columns='구분별(1)',inplace=True) # 컬럼 삭제
    dt.rename(columns={'구분별(2)' : '구분'},inplace=True) #컬럼명 변경
    # dt.set_index('구분',inplace=True) #결측값 확인하고 주석처리해서 다시 돌리기 

dt_seoul = dt[dt['구분별(1)']=='서울시']   
dt_sex = dt[dt['구분별(1)']=='성별']
dt_age = dt[dt['구분별(1)']=='연령별']
dt_edu = dt[dt['구분별(1)']=='학력별']
dt_income = dt[dt['구분별(1)']=='소득별']
dt_marry = dt[dt['구분별(1)']=='혼인상태별']
dt_big = dt[dt['구분별(1)']=='지역대분류']
dt_gu = dt[dt['구분별(1)']=='지역소분류']

rdr(dt_seoul)
rdr(dt_sex)
rdr(dt_age)
rdr(dt_edu)
rdr(dt_income)
rdr(dt_marry)
rdr(dt_big)
rdr(dt_gu)
```
